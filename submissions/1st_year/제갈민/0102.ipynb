- ### **조건문 (if): 조건에 따라 길 나누기**
프로그램이 특정 조건이 맞을 때만 코드를 실행하도록 만드는 제어문입니다.

- **기본 구조:**
  ```python
  if 조건식:
      # 조건이 참(True)일 때 실행할 코드
  elif 조건식:
      # 앞의 조건이 틀리고, 이 조건이 맞을 때 실행할 코드
  else:
      # 모든 조건이 틀렸을 때 마지막으로 실행할 코드

### 조건식을 위한 판별 연산자 목록
- a == b - a와 b가 같다
- a != b - a와 b가 다르다
- a > b - a가 b보다 크다
- a < b - a가 b보다 작다
- a >= b - a가 b보다 크거나 같다(a가 b 이상이다)
- a <= b - a가 b보다 작거나 같다(a가 b 이하다)

#### **Cell 2-2: Code (기본 예시)**
```python
# [2] 조건문 기본 예시: 성적 판독기
score = int(input())

if score >= 90:
    print("A등급입니다.")
elif score >= 80:
    print("B등급입니다.")
else:
    print("C등급입니다.")

### **반복문 1 - for문: 반복 가능한 객체를 순환하며 반복하기**
`for`문은 **Iterable(반복 가능한 객체)**에서 아이템을 하나씩 꺼내어 변수에 담아 실행하는 구조입니다.

#### 💡 Iterable(이터러블, 반복 가능한 객체)이란?
'꺼낼 수 있는 요소가 여러 개 들어 있는 데이터의 종류'라고 생각하세요!
- **List:** `[1, 2, 3]` -> 숫자 1, 2, 3을 하나씩 꺼냄
- **String:** `"AI"` -> 문자 'A', 'I'를 하나씩 꺼냄
- **range():** 정해진 범위의 숫자를 주머니에 담아둠
- **range(시작, 끝):** 시작 숫자부터 '끝 숫자 직전'까지의 숫자를 생성 (경계선 인덱싱 원리!)

**기본 구조:** `for {변수} in {Iterable}:`

### ⚙️ for문은 어떻게 작동할까?
`for`문이 실행되면 컴퓨터는 내부적으로 다음과 같은 **3단계**를 반복합니다.

1. **꺼내기**: Iterable(주머니)에서 첫 번째 요소를 꺼냅니다.
2. **담기**: 꺼낸 요소를 `for`와 `in` 사이에 있는 **변수**에 저장합니다.
3. **실행**: for문 안의 코드 블록을 실행합니다.
4. **반복**: 주머니가 빌 때까지 다음 요소를 꺼내 1~3단계를 반복합니다.

**예시 코드**

````python
# range(시작, 끝) -> 시작부터 '끝-1'까지 숫자를 생성함.
print("--- 1부터 5까지 반복 ---")
for i in range(1, 6):
    print(i,"번째 실행 중...")

# 아니면 이렇게 해도 정확히 같은 결과가 나옴.
print("--- 1부터 5까지 반복 ---")
for i in range(5):
    print(i+1,"번째 실행 중...")

# 리스트 안의 요소들을 하나씩 꺼내올 수도 있음.
clubs = ["Veritas", "Semicolon", "ENTER", "Creative Game Development"]
for club in clubs:
    print("우리 학교 IT 동아리 이름:",club)
````

### ⚠️ 파이썬의 핵심: 들여쓰기(Indentation)
파이썬은 들여쓰기를 통해 코드의 '구역'을 나눕니다.

- **들여쓰기**: 콜론 다음 줄부터는 반드시 들여쓰기를 해야 합니다. (보통 **Space 4번** 또는 **Tab**)
- **들여쓰기가 끝나는 지점**: 반복문이 종료되고 다시 원래 흐름으로 돌아가는 지점입니다.

```python
for i in range(10):
    print('Hello')    # for문 안
print('World')        # for문 밖
```

### **반복문 2 - while: 조건이 만족되는 동안 계속하기**
`for`문이 '횟수' 중심이라면, `while`문은 '조건' 중심입니다. 
`while`문은 무작정 반복하는 것이 아니라, **특정 조건이 참(True)일 때만** 작동합니다. 즉, `while`문 안에 `if`문의 논리가 이미 들어있는 셈이죠!

- **작동 원리:** "조건이 맞니?(`if`처럼 판단) -> 맞으면 실행! -> 다시 올라가서 또 물어보기!"
- **비교:** `for`문이 정해진 횟수를 돌린다면, `while`문은 **조건이 깨질 때까지** 끝없이 달립니다.

````python
# [3] while문: 나무를 10번 찍는 과정
# (조건): tree_hit이 10보다 작은 동안 반복해라!

tree_hit = 0

while tree_hit < 10:
    tree_hit += 1
    print(f"나무를 {tree_hit}번 찍었습니다.")

print("나무가 넘어갑니다! 🌲")
````

### 📝 [실습 2] 조건에 도달할 때까지 반복하기
`while`문은 특정 조건이 거짓(`False`)이 될 때까지 실행됩니다. 
단순히 숫자를 1씩 더해가며 100이 되면 멈추는 구조를 만들어봅시다.

````python
n = 0
while n < 1000:
    n += 1
    print(f"n에 {n}만큼 추가되었습니다.")
print("n이 1000이되었습니다!!!")
````

### 사용자 정의 함수 (def): 코드 묶어서 이름 붙이기
반복되는 코드 덩어리에 이름을 붙여서 필요할 때마다 불러 쓰는 것을 **함수**라고 합니다.

- **def**: "함수를 정의(Define)하겠다"는 선언
- **매개변수(Parameter)**: 함수가 실행될 때 필요한 재료(괄호 안에 들어감)
- **return**: 함수가 일을 끝내고 최종적으로 돌려주는 결과물

위와 같이 정의한 사용자 정의 함수는 일반 함수(`input()`, `print()` 등)처럼 사용할 수 있습니다.

**기본 구조**
```python
def 함수이름(파라미터1,파라미터2,...):
    실행할 코드
    return 리턴값   # 리턴은 선택적임
```

**예시 코드(두 수를 더해주는 함수)**

````python
def add_numbers(a, b):
    result = a + b
    return result

# 함수 사용하기 (호출)
result = add_numbers(10, 20)
print("함수 실행 결과:", result)
````

````python
a = int(input("첫번쨰 숫자를 입력하세요: "))
b = int(input("두번쨰 숫자를 입력하세요: "))

def add_numbers(a, b):
    result = a + b
    return result

result = add_numbers(a, b)
print("두 숫자를 더한 값:", result)
````

### 📝 [실습 3] 사각형의 넓이 계산 함수 만들기
가로와 세로 길이를 넣으면 넓이를 계산해서 돌려주는 함수를 작성해봅시다.

````python
width = int(input("가로의 길이를 입력하세요: "))
height = int(input("세로의 길이를 입력하세요: "))

# 여기에 코드를 작성하세요.

def 사각형의_넓이(width, height):
    result = width*height
    return result
result = 사각형의_넓이(width, height)
print("사각형의_넓이:", result)
````
